Imports

```python
import pandas as pd
import os
```

Loop through all XLSX files in the specified directory and convert them to csv

```python
input_dir = r"C:\Users\rohit\OneDrive\Documents\CS105\FinalProject\data"
for filename in os.listdir(input_dir):
    if filename.endswith('.xlsx'):
        # Read the XLSX file
        xlsx_file = pd.read_excel(os.path.join(input_dir, filename))
        
        # Convert to CSV format
        csv_file = xlsx_file.to_csv(os.path.join(input_dir, filename.replace('.xlsx', '.csv')), index=None, header=True)
        
        # Create a directory called "data"
        if not os.path.exists(os.path.join(input_dir, 'data')):
            os.makedirs(os.path.join(input_dir, 'data'))
        
        # Move the CSV file to the "data" directory
        os.replace(os.path.join(input_dir, filename.replace('.xlsx', '.csv')), os.path.join(input_dir, 'data', filename.replace('.xlsx', '.csv')))
```
